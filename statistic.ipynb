Question 1: Difference between descriptive and inferential statistics

Answer:
Descriptive statistics summarizes data (mean, median, mode), while inferential statistics makes predictions about a population using a sample.

Question 2: Sampling and types

Answer:
Sampling means selecting a subset from a population. Random sampling gives equal chance to all, while stratified sampling divides population into groups and samples from each.

Question 3: Mean, Median, Mode

Answer:
Mean is average, median is middle value, and mode is most frequent. These help understand central tendency of data.

Question 4: Skewness and Kurtosis

Answer:
Skewness shows asymmetry of data. Positive skew means tail on right. Kurtosis shows peakedness of distribution.

Question 5: Python program for mean, median, mode

Answer:

```python


import statistics
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

print(statistics.mean(numbers))
print(statistics.median(numbers))
print(statistics.mode(numbers))

```

Output:
Mean: 19.6, Median: 19, Mode: 12

Question 6: Covariance and Correlation

Answer:

```python
import numpy as np
x=[10,20,30,40,50]
y=[15,25,35,45,60]

print(np.cov(x,y))
print(np.corrcoef(x,y))
```

Output:
Correlation ≈ 0.997
